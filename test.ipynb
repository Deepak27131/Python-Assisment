```python
# list separte the number one list and decimla one list and text in one list 
list = [10, 2.5, "deepak", 7, 8.9, "python", 3]

numbers = []
decimals = []
text = []

for item in list:
    if isinstance(item, int):
        numbers.append(item)
    elif isinstance(item, float):
        decimals.append(item)
    elif isinstance(item, str):
        text.append(item)

print(numbers)
print( decimals)
print( text)


```
