```python
# 7 Given a list of numbers, replace every element with the greatest element on its right
# side.

arr = [4, 5, 2, 10, 8]

result = []

for i in range(len(arr)):
    if i == len(arr) - 1:
        result.append(-1)
    else:
        max_val = arr[i+1]
        for j in range(i+1, len(arr)):
            if arr[j] > max_val:
                max_val = arr[j]
        result.append(max_val)

print(result)
```

```python
#8 Take a list and find the longest increasing contiguous sublist.

num = [1, 2, 2, 3, 4, 1, 2, 3, 4, 5]

max_list = []
current = [num[0]]

for i in range(1, len(num)):
    if num[i] > num[i-1]:
        current.append(num[i])
    else:
        if len(current) > len(max_list):
            max_list = current
        current = [num[i]]

# last sequence check
if len(current) > len(max_list):
    max_list = current

print(max_list)
```

```python
num = [1, 2, 2, 3, 4, 1, 2, 3, 4, 5]

current = [num[0]]   # current increasing sublist
longest = [num[0]]   # longest sublist

for i in range(1, len(num)):
    if num[i] > num[i-1]:
        current.append(num[i])   # increase ho raha hai → add karo
    else:
        current = [num[i]]       # break ho gaya → reset

    if len(current) > len(longest):
        longest = current        # update longest

print(longest)
```
