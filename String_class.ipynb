```python
text = "There is a vovel or counsunent"
count = 0

for i in range(len(text)):
    if text[i].lower() in ['a', 'e', 'i', 'o', 'u']:
        count += 1

print("Number of vowels:", count)
```

```python
# Check whether a given string is a palindrome or not.

a = 'madam'
reverse = a[::-1]

if reverse == a:
    print("It is a palindrone  string: ", a)
else :
    print("Not a palindrone: ",a)
```

```python
def is_palindrome(n):
    num = n
    result = 0

    while num > 0:
        Id = num % 10
        result = result * 10 + Id
        num = num // 10

    return n == result


print(is_palindrome(5873))
```
