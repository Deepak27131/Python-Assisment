## DICTIONARY

In Python, a dictionary is an unordered, mutable collection that stores data in key–value pairs. Keys must be unique and immutable (like strings, numbers, tuples), while values can be of any type. Dictionaries are one of the most powerful data structures in Python for fast lookups and flexible data storage.

Method	Description	Example

1. clear()	Removes all items	my_dict.clear() → {}

2. copy()	Returns a shallow copy	new_dict = my_dict.copy()

3. fromkeys(seq, val)	Creates dict from sequence	dict.fromkeys([1,2,3], 0) → {1:0,2:0,3:0}

4. get(key, default)	Returns value if key exists, else default	my_dict.get("age", 0)

5. items()	Returns key–value pairs as tuples	my_dict.items() → [("name","Deepak"),("age",25)]

6. keys()	Returns all keys	my_dict.keys()

7. values()	Returns all values	my_dict.values()

8. pop(key)	Removes item with given key	my_dict.pop("age")


9. popitem()	Removes last inserted item	my_dict.popitem()

10. setdefault(key, val)	Returns value if key exists, else inserts with val	my_dict.setdefault("city","Jaipur")

11. update(dict2)	Updates dictionary with another	my_dict.update({"age":26})

```python
# 1. Count frequency of each character in a string using a dictionary.
s= 'pineapple'
d = {}
for ch in s:
    if ch not in d:
        d[ch] = 1
    else:
        d[ch] += 1      
d
       
    
```

```python
# 2. Count frequency of each word in a sentence.
sentence = "Count frequency of each word in a sentence"

words = sentence.split()

freq = {}

for ch in words:
    freq[ch] = freq.get(ch,0)+1
    
print(freq)
```

```python
# 5. Sort a dictionary by its values.

student_marks = {
    "Aman": 85,
    "Riya": 92,
    "Kunal": 78
}

items = list(student_marks.items())

for i in range(len(items)):
    for j in range(i+1, len(items)):
        if items[i][1] > items[j][1]:
            items[i],items[j] = items[j], items[i]
            
sorted_dict = dict(items)
print(items)
print(sorted_dict) 
```

```python
#11. Group elements of a list into a dictionary based on their first letter.

list = ["apple", "banana", "apricot", "cherry", "blueberry"]

grouped = {}

for i in list:
    fist = i[0]
    if fist not in grouped:
        grouped[fist] = []
    grouped[fist].append(i)
    
print(grouped)
    


```

```python
# 14. Find duplicate values in a dictionary.
d = {"a": 10, "b": 20, "c": 30, "d": 30}

find = {}

for value in d.values():
    if value in find:
        print("duplicate value", value)
    else:
        find[value] = 1
```

```python
d = {"a": 10, "b": 20, "c": 30, "d": 30}

find = {}
duplicates = []

for value in d.values():
    if value in find:
        if value not in duplicates:
            duplicates.append(value)
    else:
        find[value] = 1

print(duplicates)
```

```python
# 15. Flatten a nested dictionary.
student = {
    "Aman": {"age": 20, "city": "Delhi"},
    "Riya": {"age": 22, "city": "Mumbai"}
}

flattened = {}

for outer_key, inner_dict in student.items():
    for inner_key, value in inner_dict.items():
        flattened[outer_key + "_" + inner_key] = value

print(flattened)
```

```python
nested_dict = {
    "Aman": {"age": 20, "city": "Delhi"},
    "Riya": {"age": 22, "city": "Mumbai"}
}

flattened = {}

for outer_key, inner_dict in nested_dict.items():
    for inner_key, value in inner_dict.items():
        # nayi key banai: outerkey + "_" + innerkey
        new_key = f"{outer_key}_{inner_key}"
        flattened[new_key] = value

print(flattened)
```
