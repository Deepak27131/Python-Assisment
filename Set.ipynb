```python
# 10. Find common elements between two lists using sets.
list = [1,2,3,4,5,6,7]
list1 = [2,2,1,4,56,3,7]

sets1 = set(list)
sets2 = set(list1)
sets1.intersection(sets2)
```

```python
# 12. Find elements present in one list but not in another.

list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]


set1 = set(list1)
set2 = set(list2)

diff1 = set1 - set2
print("In list1 but not in list2:", diff1)

# Elements in list2 but not in list1
diff2 = set2 - set1
print("In list2 but not in list1:", diff2)
```

```python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

diff = set1.difference(set2)
print(diff)  
```

```python
# 13. Check if a set is empty.
sets = set()
# sets = {1, 2, 3}

for i in sets:
    print("Not empty")
    break
else:
    print("Empty")
```

```python
my_set = set()   # empty set

if not my_set:
    print("Set is empty")
else:
    print("Set is not empty")
```

```python
my_set = {1, 2, 3}

if len(my_set) == 0:
    print("Set is empty")
else:
    print("Set is not empty")
```

```python
# 14. Create a set of squares from 1 to N.

N = 10
squares = set()

for i in range(1, N+1):
    squares.add(i**2)

print(squares)
```

```python
my_set = {1, 2, 3, 4, 5}
squared_set = set()

for i in my_set:
    squared_set.add(i**2)

print(squared_set)
```

```python
N = 10

# Using set comprehension
squares = {x**2 for x in range(1, N+1)}

print(squares)
```

```python

#  15. Count unique elements in a list using a set.


my_list = [1, 2, 2, 3, 4, 4, 5, 1]


unique_elements = set(my_list)


count = len(unique_elements)

print("Unique elements:", unique_elements)
print("Count:", count)
```
