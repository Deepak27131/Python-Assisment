```python
# unions 
A = {1,2,3,4}
B = (3,4,5,6)
A.union(B)
```

```python
# intersection 
A.intersection(B)
```

```python
# Symmetric Differemnce 
A.symmetric_difference(B)
```

```python
s1 = {1, 2, 3}
s2 = {3, 4, 5}

print(s1.union(s2))              # {1,2,3,4,5}
print(s1.intersection(s2))       # {3}
print(s1.difference(s2))         # {1,2}
print(s1.symmetric_difference(s2)) # {1,2,4,5}
```
