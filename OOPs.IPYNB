# 1. What is OOP?
Object-Oriented Programming is a way of organizing code into Classes (blueprints) and Objects (instances of those blueprints). It helps make code reusable and easier to manage. 
2. The Core Method: __init__
The most important method in Python OOP is the constructor, called __init__. It runs automatically when you create a new object and sets up its initial data. 
3. The 4 Pillars of OOP
Encapsulation: Keeping data (variables) and methods together in one unit (class) and restricting access to some details.
Inheritance: Letting one class "borrow" features from another class.
Polymorphism: Allowing different classes to use the same method name but behave differently.
Abstraction: Hiding complex logic and only showing the necessary "interface" to the user. 

## 1. Class: The Blueprint
A Class is a logical entity. It is a user-defined blueprint or template from which objects are created. It defines a set of attributes (variables) and methods (functions) that the objects of that class will have.
Analogy: Think of a class as a House Map. The map itself is not a house; it just describes where the walls, doors, and windows will be.
Key Fact: A class does not occupy memory space until an object is created.
2. Object: The Instance
An Object is a physical entity. It is a specific instance of a class. When a class is defined, no memory is allocated until an object of that class is created. Each object has its own state (data) and behavior.
Analogy: If the map is the Class, the Actual House built using that map is the Object. You can build 10 different houses (objects) from the same map (class).
Key Fact: Objects occupy memory and can interact with each other.
Proper Definitions
Term	Definition
Class	A collection of objects is called a Class. It is a logical entity that contains methods and variables common to all instances.
Object	An instance of a class that has a state and behavior. It is the basic runtime entity of an Object-Oriented system.
Detailed Code Example
python
### Defining the Class
class Smartphone:
    # The Constructor (sets up the object's initial data)
    def __init__(self, brand, model, storage):
        self.brand = brand      # Attribute
        self.model = model      # Attribute
        self.storage = storage  # Attribute

    # A Method (Action/Behavior)
    def description(self):
        return f"This is a {self.brand} {self.model} with {self.storage}GB storage."

### Creating Objects (Instantiating the class)
phone1 = Smartphone("Apple", "iPhone 15", 128)
phone2 = Smartphone("Samsung", "S24 Ultra", 256)

#### Accessing Object Properties and Methods
print(phone1.description()) 
##### Output: This is a Apple iPhone 15 with 128GB storage.

print(phone2.brand) 
##### Output: Samsung
Use code with caution.
Summary of Differences
Definition: A class is a template; an object is the implementation of that template.
Memory: Class is a logical structure (no memory); Object is a physical entity (allocated memory).
Existence: You define a class once, but you can create as many objects from it as you need.

### Polymorphism
Polymorphism ek Object-Oriented Programming concept hai jisme same interface (method/function/operator) different objects ke liye different behavior show karta hai. Yeh abstraction ko support karta hai aur allow karta hai ki ek hi naam ka function multiple forms me execute ho depending on object type, runtime context, ya input parameters.

* . Types of Polymorphism (Full Deep Breakdown)

1. Compile-Time Polymorphism (Static Binding)

    compile-time polymorphism me function ka behavior compile hone ke time decide hota hai. Iska main concept hai method overloading, jisme same function name ke multiple versions hote hain with different parameters. Python me direct overloading nahi hota, but default arguments ya variable-length arguments se simulate kiya jata hai.

2. Run-Time Polymorphism (Dynamic Binding)

    Run-time polymorphism me method ka execution runtime pe decide hota hai, na ki compile time pe. Yeh mostly method overriding ke through achieve hota hai jahan child class parent ke method ko redefine karti hai. Jab object call hota hai, tab actual method execution decide hota hai based on object type.

```python
# Encapsulation

class ATM:
    def __init__(self,pin,card_holder,cvv):
        self.__pin = pin # Private attribute(__)
        self.card_holder = card_holder # public attribute
        self._cvv = cvv  # protected attribute (_)
        
card = ATM(123,'Deepak Kumar',243)
print(card._ATM__pin) # name magling is the process of aceesing the private attributes Private 
# syntax --> obj._ATM__PIN
print(card.card_holder)  # public attribute
print(card._cvv)  # protected instance attribute
```

```python
class ATM:
    def __init__(self , pin, card_holder):
        self.__pin = pin
        self.card_holder = card_holder
      
    # setter method
    # this method is used into the encapsulation the set the new value to private and protected attributes
      
    def change_pin(self,new_pin):
        self.__pin = new_pin
        print("pin is change succesfully")
    
    # getter method --> this method is used to get the hidden attributes outside the class 
    def checking_pin(self):
        print(self.__pin)        
    

card = ATM(8643,'deepak')

card.change_pin(123)
card.checking_pin()

```

```python
#  inheritance 

class Human:
    def __init__(self , name, age , location):
        self.name = name
        self.age = age 
        self.location = location
        
class Student(Human):
    def __init__(self ,name ,age , location,course):
        super().__init__(name,age , location) # by using the super keyword or the method we access the parents class attributes and initaliszation
        # by calling the explicitely the init method
        self.course = course
    

obj = Student('Rahul', 22,'deoria','B.tech')
print(obj.name)
print(obj.course)
```

```python
# abstraction 
from abc import ABC, abstractmethod

# Abstract Class
class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass

# Child Class
class Car(Vehicle):
    def start(self):
        print("Car starts with key")

class Bike(Vehicle):
    def start(self):
        print("Bike starts with kick")

# Objects
c = Car()
b = Bike()

c.start()
b.start()
```

```python
# classmethod -->Recive the class as the fisrt arguments(cls)
# can acces and modify class level attribute via cls

class Student:
    college = 'IIT'
    def __init__(self ,name):
        self.name = name
        
    @classmethod
    def change_college(cls,new_name):
        cls.college = new_name
        
s = Student('Rahul')
s.college
s.change_college('ITI')
s.college
```
