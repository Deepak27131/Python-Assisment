# 🔹 What is a Function in Python?

1. In Python, a function is a reusable block of code that performs a specific task. Functions help avoid repetition, make code modular, and can accept inputs (parameters) and return outputs (values)

Function is a block of code which is define to perform a specific task. It is reusable and can be called multiple times in a program. Functions can take parameters (inputs) and can return values (outputs). They help in organizing code, improving readability, and following the DRY (Don't Repeat Yourself) principle.

2. A function is a named block of code that runs only when called.

3. It can take arguments (inputs) and can return a value (output).

4. Functions follow the DRY principle (Don’t Repeat Yourself), meaning you write code once and reuse it multiple times.

5. basically when we define function then  using the keyword 'def' we basically give two ar more variable they are also called as 'paramaters'.
And when we call the function usimng function name then you can pass some values into the function and this value are called as 'arguments'


### 🔹 Why Functions are Important

Reusability – Write once, use many times.

Modularity – Break big problems into smaller parts.

Readability – Code becomes cleaner and easier to understand.

Maintainability – Easier to debug and update.

#### 🔹 2. Built-in Functions

These are functions that Python already provides. You don’t need to define them; just call them directly.

📌 Categories of Built-in Functions

* Input/Output Functions → print(), input()

* Type Conversion Functions → int(), float(), str(), list()

* Mathematical Functions → abs(), pow(), round(), min(), max()

* Sequence Functions → len(), sum(), sorted(), reversed()

* Utility Functions → type(), id(), help(), dir()

```python
import re

states = ["  Alabama","Geogia!","Georgia","FlorID","sounth  carolinaa##","west  virignia?"]

def remove_punctuation(value):
    return re.sub("[!?#]", "", value)

clean_ops = [str.strip, remove_punctuation, str.title]

def clean_string(strings, ops):
    result = []
    for value in strings:
        for func in ops:
            value = func(value)
        result.append(value)
    return result

print(clean_string(states, clean_ops))
```

## High oder fuction built in the python

1. Map --> map():  map function is used to map any function on any of the iterable which maps all the element with that function.
Map fuction is a high order fuction which maps all the element of the iterable at the same time.
it return at the same time.

Syntax --> map(function,iterable(li,tuple))

2. Filter Function()--> filter(function,iterable):  is a high order fuction which take  onother function as argument which filter out the element of the iterable those satisefied that condition  only TRUE element have extract filter().
Synttax--> ilter(function,iterable(list , tuple))




```python
# map function
li =[ 1,2,3,4,5]
def square(num):
    return num * num

ans = list(map(square,li))
ans
```

```python
# iterable function

li =[ 1,2,3,4,5]
def is_even(num):
    if num % 2 == 0:
        return True
    return False

ans = list(filter(is_even,li))
ans
```

```python
# iterable function

li =[ 1,2,3,4,5]
def is_even(num):
    return num % 2 == 0
       

ans = list(filter(is_even,li))
ans
```
