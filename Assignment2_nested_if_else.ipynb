```python
# 1. A student enters marks. If marks are greater than or equal to 40, then check if marks
#are greater than 75 and print “Distinction”; otherwise print “Pass”. If marks are below
#40, print “Fail”.  

marks = int(input("Enter marks: "))

if marks >= 40:
    if marks > 75:
        print("Distinction")
    else:
        print("Pass")
else:
    print("Fail")
```

```python
# 2. Take a number from the user. First check whether it is positive. If it is positive, then
#check whether it is even or odd and print the result.


number = int(input("Enter a number: "))

if number > 0:
    print("The number is positive.")
    if number % 2 == 0:
        print("The number is even.")
    else:
        print("The number is odd.")
else:
    print("The number is not positive.")
```

```python
# 3. A person enters their age. If age is 18 or above, then check whether age is above 60
# to classify them as a senior citizen; otherwise classify them as an adult. If age is below
# 18, print “Minor”.

age = int(input("Enter your age: "))

if age >= 18:
    print("Age is greater than or equal to 18.")
    if age > 60:
        print("Senior Citizen")
    else:
        print("Adult")
else:
    print("Minor")
```

```python
# 4. A company checks employee eligibility. If experience is greater than 2 years, then
# check if salary is below 30,000 to approve a salary hike; otherwise no hike. If experience is 2 years or less, print “Not eligible for hike”

experience = int(input("Enter years of experience: "))

if experience > 2 :
    print("Experience is greater than 2 years.")
    salary = int(input("Enter your salary: "))
    if salary < 30000:
        print("Salary hike approved.")
    else:
        print("No hike.")
else:
    print("Not eligible for hike.")

```

```python
# 5. Take a number as input. If it is divisible by 2, then check whether it is also divisible
# by 4 and print the appropriate message.

number = int(input("Enter a number: "))

if number % 2 == 0:
    print("The number is divisible by 2.")
    if number % 4 == 0:
        print("The number is also divisible by 4.")
    else:
        print("The number is not divisible by 4.")
        
else:
    print("The number is not divisible by 2.")
```

```python
# 16. A movie theater checks age. If age is above 13, then check whether the show is
# rated 18+ to allow or deny entry.

age = int(input("Enter your age: "))
show_rating = input("Enter show rating (U/A/12A/15/18): ")

if age > 13:
    print("Age is above 13.")
    if show_rating == "18" or show_rating == "18+":
        print("Entry allowed.")
    else:
        print("Entry allowed.")
else:
    print("Entry denied. You must be above 13 years old.")
```

```python
# 17. Take a number. If it is a three-digit number, then check whether all digits are equal
# and print the result.

number = int(input("Enter a number: "))

if 100 <= number <= 999:
    print("The number is a three-digit number.")
    first_digit = number // 100
    second_digit = (number // 10) % 10
    third_digit = number % 10
    
    if first_digit == second_digit == third_digit:
        print("All digits are equal.")
    else:
        print("All digits are not equal.")
else:
    print("The number is not a three-digit number.")
```

```python
# 18. A gym checks membership. If membership exists, then check payment status
# before allowing entry.

membership_exists = input("Do you have membership? (Yes/No): ").lower()

if membership_exists == "yes":
    print("Membership exists.")
    payment_status = input("Is payment done? (Yes/No): ").lower()
    
    if payment_status == "yes":
        print("Payment verified. Entry allowed.")
    else:
        print("Payment pending. Entry denied.")
else:
    print("No membership. Entry denied.")
```

```python
# 22. A user enters password. If password length is valid, then check whether it contains
# a number and print password strength.

password = input("Enter password: ")

if len(password) >= 8:
    print("Password length is valid.")
    has_number = False
    
    for char in password:
        if char.isdigit():
            has_number = True
            break
    
    if has_number:
        print("Password contains a number. Password strength: Strong.")
    else:
        print("Password does not contain a number. Password strength: Weak.")
else:
    print("Password length is not valid. Minimum 8 characters required.")
```

```python
# 23. A company checks job applications. If qualification is valid, then check experience
# requirement before shortlisting.

qualification = input("Enter qualification (Bachelor/Master/PhD): ")

if qualification in ["Bachelor", "Master", "PhD"]:
    print("Qualification is valid.")
    experience = int(input("Enter years of experience: "))
    
    if experience >= 2:
        print("Experience requirement met. Application shortlisted.")
    else:
        print("Insufficient experience. Application rejected.")
else:
    print("Invalid qualification. Application rejected.")
```

```python
# 24. A library checks student ID. If ID is valid, then check fine amount before issuing a
# book.

student_id = input("Enter student ID: ")

if student_id.isdigit() and len(student_id) == 6:
    print("Student ID is valid.")
    fine_amount = int(input("Enter fine amount: "))
    
    if fine_amount == 0:
        print("No pending fine. Book issued successfully.")
    else:
        print(f"Pending fine of {fine_amount} rupees. Clear fine before issuing book.")
else:
    print("Invalid student ID. Book not issued.")
```

```python
# 25. A restaurant checks reservation status. If reservation exists, then check arrival time
# to allow seating.

reservation_exists = input("Is reservation made? (Yes/No): ").lower()

if reservation_exists == "yes":
    print("Reservation exists.")
    arrival_time = input("Enter arrival time (HH:MM): ")
    
    print(f"Checking arrival time: {arrival_time}")
    if arrival_time:
        print("Seating allowed. Welcome to the restaurant.")
    else:
        print("Invalid arrival time. Please wait.")
else:
    print("No reservation found. Cannot allow seating.")
```
