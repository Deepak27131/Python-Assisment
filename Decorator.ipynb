# Decorator in Python
1. decorator are the high order function in which enhance the functionality of the function wiothout changing original code.

2. A decorator takes a function as input, adds some logic, and returns a new function.

They are often used for logging, authentication, timing, caching, and validation.

Syntax uses the @decorator_name notation placed above the function definition.


```python
def my_decorator(func):
    def wrapper():
        print("Before the function runs")
        func()
        print("After the function runs")
    return wrapper

@my_decorator
def say_hello():
    print("Hello, Deepak!")

say_hello()
```

```python
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print(f"Running {func.__name__}...")
        result = func(*args, **kwargs)
        print(f"Finished {func.__name__}")
        return result
    return wrapper

@my_decorator
def greet(name):
    print(f"Hello, {name}!")

@my_decorator
def add(a, b):
    print(f"Sum is {a + b}")
    return a + b

@my_decorator
def multiply(a, b):
    print(f"Product is {a * b}")
    return a * b

# Calling multiple decorated functions
greet("Deepak")
add(5, 3)
multiply(4, 6)
```

# What is an Iterator?

1. An iterator is an object used to traverse through a collection of data one element at a time.
iterator is a statefull object meaning it remenber it current position during iteration

Iterator ek object hai jo ek sequence (list, tuple, string, etc.) ke elements ko ek-ek karke return karta hai.

Python me iterator banane ke liye iter() aur next() functions use hote hain.

Iterator protocol:

__iter__() → returns the iterator object itself.

__next__() → returns the next element, and raises StopIteration when no items are left.


### what is difference btw iterable and iterator

1. Definition: An object that can be looped over (like list, tuple, dict, set, string).

Protocol: Must implement __iter__() method, which returns an iterator.

Example:
                numbers = [1, 2, 3]   # list is iterable
                for n in numbers:
                    print(n)

2. Iterator
Definition: An object that produces elements one at a time using next().

Protocol: Must implement both __iter__() and __next__() methods.

Example:    
            numbers = [1, 2, 3]
            it = iter(numbers)    # convert iterable to iterator
            print(next(it))       # 1
            print(next(it))       # 2
            print(next(it))       # 3


```python
# A list
numbers = [10, 20, 30]

# Convert list into iterator
iterator_obj = iter(numbers)

print(next(iterator_obj))  # 10
print(next(iterator_obj))  # 20
print(next(iterator_obj))  # 30
# Next call will raise StopIteration
```

```python
class CountDown:
    def __init__(self, start):
        self.num = start
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.num <= 0:
            raise StopIteration
        current = self.num
        self.num -= 1
        return current

# Using custom iterator
for val in CountDown(5):
    print(val)
```

```python
#1. Create a class that prints numbers from 1 to 5 using iterator.

class Number:
    def __iter__(self):
        self.num = 1
        return self
    
    def __next__(self):
        if self.num > 5:
            raise StopIteration
        val = self.num
        self.num += 1
        return val
obj = Number()

for n in obj:
    print(n)
    
```

### what is  Generator 

Genrator in python are a speial type of function that allow to create iterator in memory -efficeint way. Unlike regular function that return a single value and ternimate genartor 'yield' a series of value once at a time , pausing their excecution and saving their satate between each one.

it used keyword 'yeild'

📌 Definition
A generator is a function that uses yield to return values lazily (on demand), rather than computing and storing them all at once.

📌 Key Points
yield keyword → pauses the function and returns a value, but keeps the function’s state for the next call.

Memory efficient → doesn’t store all results in memory.

Lazy evaluation → values are generated only when requested.

Iterator protocol → Generators can be used in loops (for) or with functions like next().

def simple_gen():
    yield 1
    yield 2
    yield 3

gen = simple_gen()
print(next(gen))  # 1
print(next(gen))  # 2
print(next(gen))  # 3

* 📌 Why Generators Matter
Efficient for large datasets (e.g., reading big files line by line).

Useful for streaming data.

Common in interview questions to test understanding of iterators and lazy evaluation.

### 📌 What is Exception Handling?
Exception handling in Python is a mechanism to detect and manage errors during program execution. Instead of letting the program crash when something goes wrong, Python allows you to "catch" the error and decide what to do next.

* 📌 Key Terms
Exception → An error that occurs during program execution (e.g., ZeroDivisionError, ValueError).

Handling → Writing code that responds to the error without stopping the program.

try:
    # Code that might cause an error
except ExceptionType:
    # Code that runs if that specific error occurs
else:
    # Code that runs if no error occurs
finally:
    # Code that always runs (cleanup)

* Why It Matters
Prevents program crashes.

Makes code more robust and user-friendly.

Essential in file handling, database operations, and APIs.

* There are 4 type ;

1. Try --> try block alway run to check there any exception or not 
2. except --> if the exception has occured into this then the exception will run
3. else --> if the try block an succesfully then only the else block will excetive
4. finally --> finally block will run always 

5. Custom Exception --> are the exception which are made by coder (user define) into that we basically raise the custome exception using 'raise' keywword


```python
# generator 
def normal():
    li = [1,2,3,4,5]
    for el in li:
        yield el

ans = normal()
print(next(ans))
print(next(ans))
print(next(ans))
print(next(ans))
```

```python
# 12. Write a function that raises an exception if input is not a string.

def check_string_input(x):
    try:
        if not isinstance(x, str):
            raise TypeError("Input is not a string!")
    except TypeError as e:
        print("Error:", e)
    else:
        print("Valid string:", x)

# Example calls
check_string_input("Hello")
check_string_input(123)   # will raise exception
```
