```python
# WAP to check whether numbers up to n are prime or not

n = 10
i = 2   # start from 2, since 1 is not prime

while i <= n:
    j = 2
    is_prime = True
    
    while j <= i // 2:   # check divisors up to half of i
        if i % j == 0:
            is_prime = False
            break
        j += 1
    
    if is_prime:
        print("Prime Number:", i)
    else:
        print("Not a prime number:", i)
    
    i += 1
```
