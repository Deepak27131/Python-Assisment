```python
# codition 1
# left diagonal  star (n-i +1) == col 
# right diagonal star  (n+i - 1) == col
# bootom row --> i == n

n = 4
for i in range(1,n+1):
    for col in range(1,2*n):
        if (n - i +1) == col or (n+i-1) == col or i == n:
            print("*" , end=" ")
        else:
            print(" " ,end =" ")
    print()
```
