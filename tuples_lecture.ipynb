## TUPLE IN PYTHON 

1. Tuple is data type which is immutable in nature and tuple hold the sequence of data 
2. tuple is faster than list beacuse their immutable allow the python interpreter to make memory and performance optimization that are not possible with utalbe list

    
    The key reasons for the performance difference are:

        Memory Efficiency & Allocation:
        Tuples are stored in a single, contiguous block of memory and do not over-allocate space because their size is fixed upon creation.
        Lists are mutable and must over-allocate memory to efficiently handle future operations like append(), which avoids expensive reallocations every time an item is added. This overallocation makes lists larger and slightly less memory efficient.

3. Tuple Packing --> is the proces of  grouping multiple values into a single tuple variable in python.
This feature is one of the things than makes python mutiple assignment syntax possible and elegant

4. Tuple Unpacking --> also knowns as mutiple assigmenet or iterable unpacking is a python feature that allow to  you extract individual value from a tuple (or any other iterable like a list ,string or dictionary)
 and assign them to muttiple variable in asingle statement

```python
# 14. Sort a tuple.
li = (1, 0, 2, 2, 3, 4, 3, 4, 5, 6)

temp = list(li)

for i in range(len(temp)):
    for j in range(i + 1, len(temp)):
        if temp[i] > temp[j]:
            temp[i], temp[j] = temp[j], temp[i]

sorted_tuple = tuple(temp)

print(sorted_tuple)
```

```python
li = (1, 0, 2, 2, 3, 4, 3, 4, 5, 6)

sorted_tuple = tuple(sorted(li))

print(sorted_tuple)
```
